```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
storesdata = pd.read_csv("Stores.csv")
```

```python
storesdata.describe() #basic stats 
```

```python
f, axes = plt.subplots(4, 1, figsize=(16, 8))

sb.histplot(data = storesdata[["Store_Area"]], ax = axes[0])
sb.histplot(data = storesdata[["Items_Available"]], ax = axes[1])
sb.histplot(data = storesdata[["Daily_Customer_Count"]], ax = axes[2])
sb.histplot(data = storesdata[["Store_Sales"]], ax = axes[3])
```

```python
f, axes = plt.subplots(4, 1, figsize=(16, 8))

sb.boxplot(data = storesdata[["Store_Area"]], ax = axes[0], orient = "h")
sb.boxplot(data = storesdata[["Items_Available"]], ax = axes[1], orient = "h")
sb.boxplot(data = storesdata[["Daily_Customer_Count"]], ax = axes[2], orient = "h")
sb.boxplot(data = storesdata[["Store_Sales"]], ax = axes[3], orient = "h")
```

```python
storesdata_sorted_by_sales = storesdata.sort_values(by=['Store_Sales'])
storesdata_sorted_by_sales
```

```python
f = plt.figure(figsize=(16, 8))

plt.scatter(storesdata[["Store_Sales"]], storesdata[["Daily_Customer_Count"]])
plt.scatter(storesdata[["Store_Sales"]], storesdata[["Store_Area"]])
plt.scatter(storesdata[["Store_Sales"]], storesdata[["Items_Available"]])
```

```python
f = plt.figure(figsize=(16, 8))

plt.scatter(storesdata[["Daily_Customer_Count"]], storesdata[["Store_Sales"]])
plt.scatter(storesdata[["Daily_Customer_Count"]], storesdata[["Store_Area"]])
plt.scatter(storesdata[["Daily_Customer_Count"]], storesdata[["Items_Available"]])
```
